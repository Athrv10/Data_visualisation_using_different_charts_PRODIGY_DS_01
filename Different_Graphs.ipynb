## **DATA VISUALIZATION**

```python
data1 = data.sort_values(by='population_total',ascending=False).head(10)
```

```python
sns.barplot(x=data1['name_of_city'],y=data1['population_total'],data=data)
plt.title("Top 10 city popoulation")
plt.xlabel("City")
plt.ylabel("Population_total")
plt.show()
```

```python
data2 = data.sort_values(by='population_total',ascending=False).tail(10)
```

```python
sns.barplot(y=data2['name_of_city'],x=data2['population_total'],data=data)
plt.title("Top 10 city popoulation")
plt.xlabel("City")
plt.ylabel("Population_total")
plt.show()
```

```python
sns.barplot(x=['male','female'],y= [data.population_male.sum() , data.population_female.sum()])
plt.title("Total Male and Female Population")
plt.xlabel("Gender")
plt.ylabel("Population")
plt.show()
```

```python
sns.barplot(x=['Male','Female'],y= [data.literates_male.sum(),data.literates_female.sum()])
plt.title("Total Litereates")
plt.xlabel("Gender")
plt.ylabel("Population")
plt.show()
```

```python
top_5_cities = data.sort_values(by='population_total', ascending=False).head(5)

# Melt the DataFrame to create a long-form dataset for seaborn
melted_data = top_5_cities.melt(
    id_vars=['name_of_city'],
    value_vars=['population_male', 'population_female'],
    var_name='gender',
    value_name='population'
)

# Replace gender variable values for better readability in the plot
melted_data['gender'] = melted_data['gender'].replace({'population_male': 'Male', 'population_female': 'Female'})

# Create the grouped bar plot using seaborn
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.barplot(x='name_of_city', y='population', hue='gender', data=melted_data)
plt.title('Male and Female Population Distribution in Top 5 Cities')
plt.xlabel('City')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['child_sex_ratio'], bins=20)
plt.title('Distribution of Child Sex Ratio')
plt.show()
```

```python
import random


def create_random_bar_chart(data, x_col, y_col, title):
    random_subset = data.sample(n=min(10, len(data)))
    plt.figure(figsize=(8, 6))
    sns.barplot(x=x_col, y=y_col, data=random_subset)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


for i in range(10):
    x_col = random.choice(data.columns)

    numeric_cols = data.select_dtypes(include=np.number).columns.tolist()

    if numeric_cols:
        y_col = random.choice(numeric_cols)
        title = f" {i+1}: {x_col} vs. {y_col}"
        try:
            create_random_bar_chart(data, x_col, y_col, title)
        except Exception as e:
            print(f"Error creating chart {i+1}: {e}")

    else:
        print(f" {i + 1}")
```
